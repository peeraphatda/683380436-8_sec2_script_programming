```python
# Snippet A
score = 85
if score >= 90:
grade = "A";
elif score >= 80:
grade = "B";
else:
grade = "C";
print(f;"Grade: {grade}";)
# What's the output?
```

Ans : โจทย์กำหนดค่าของ score เป็น 85 และหากค่าของ score น้อยกว่าหรือเท่ากับ 90 ซึ่งเป็นค่ากำหนดของ if grade จะถูกประกาศเป็น A ตามคำสั่งของ print แต่หากค่าของ score น้อยกว่าหรือเท่ากับ 80 ซึ่งเป็นค่ากำหนดของ elif gradeจะถูกประกาศเป็น B นอกเหนือจากนั้น grade จะถูกประกาศเป็น C และ เนื่องจาก score = 85 ดังนั้น output คือ  "Grade: A"

```python
# Snippet B
temp = 25
is_raining = True
if temp > 20 and not is_raining:
print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
print("Stay indoors or bring an umbrella.")
else:
print("Unsure about the weather.")
# What's the output?
```

Ans :โจทย์กำหนดค่าของ temp เป็น 25 และ is_raining เป็น true โดยเงื่อนไขของ if แรกถูกกำหนดไว้ว่าจะต้องมีค่า temp มากกว่า 20 และ is_raining ต้องเป็น false หากตรงตามเงื่อนไขจะแสดงข้อความว่า "Perfect day for a picnic!"
elif ถูกกำหนดไว้ว่า temp น้อยกว่าหรือเท่ากับ 20 หรือ is_raining ต้องเป็น true
หากตรงตามเงื่อนไขจะแสดงข้อความว่า
"Stay indoors or bring an umbrella."
ในกรณีนอกเหนือจากข้างต้น จะแสดงข้อความว่า "Unsure about the weather."
ซึ่งในกรณีของโจทย์ค่าที่กำหนดมาคือ temp เป็น 25 และ is_raining เป็น true
ดังนั้น output ของโจทย์นี้คือ "Stay indoors or bring an umbrella."

```python
# Snippet C
age = 17
has_license = True
if age >= 18:
if has_license:
print("You can drive legally.")
else:
print("You need a license to drive.")
else:
print("You are too young to drive.")
# What's the output?
```

Ans :โจทย์กำหนด age เป็น 17 และ has_license เป็น True if แรกถูกกำหนดค่าไว้ว่าต้องมี age มากกว่าหรือเท่ากับ 18 หากเข้าเงื่อนไขจะพาไปยัง if ถัดไป โดย if สอง ถูกกำหนดไว้ว่าหาก has_license เป็น True จะแสดงข้อความ "You can drive legally."
หากใน if 2 เงื่อนไขไม่ตรงและเป็นอย่างอื่นนอกจากนั้นข้อความจะแสดง "You need a license to drive."
และหากไม่ตรงเงื่อนไขใน if แรก จะแสดงข้อความ "You are too young to drive."
โดยในโจทย์นี้โจทย์กำหนดไว้ว่า age เป็น 17 และ has_license เป็น True ดังนั้น output ของโจทย์นี้คือ "You are too young to drive."
