```python
print("=== ยินดีต้อนรับสู่เกมหลงป่าอาถรรพ์ ===")
path = input("คุณเดินมาเจอทางแยก จะไปทาง 'ซ้าย' (ไปแม่น้ำ) หรือ 'ขวา' (เข้าป่าลึก)? ")

if path == "ซ้าย":
    print("คุณเดินมาถึงริมแม่น้ำที่เชี่ยวกราก และเจอเรือพายเก่า ๆ ลำหนึ่ง")
    boat_choice = input("จะ 'พาย' เรือข้ามไป หรือ 'เดิน' เลาะตลิ่ง? ")

    if boat_choice == "พาย":
        print("เรือพายพารอดฝั่งได้อย่างปลอดภัย! คุณเจอหมู่บ้านและชนะเกมนี้!")
    elif boat_choice == "เดิน":
        print("แย่แล้ว! เดินเลาะตลิ่งไปเจอเสือโคร่งเจ้าป่า คุณต้องวิ่งหนีกลับมาที่เดิม")
    else:
        print("ตัวเลือกไม่ถูกต้อง เกมจบลงเพราะคุณยืนลังเลจนค่ำ")

elif path == "ขวา":
    print("คุณเดินลึกเข้าไปในป่าที่มืดมิดและไร้ทางออก... คุณหลงป่า!")

else:
    # ส่วนของ Responsible AI: ป้องกันการใส่คำสั่งอื่นที่ระบบไม่รองรับ
    print("ระบบไม่เข้าใจคำสั่งของคุณ (กรุณาเลือกพิมพ์ 'ซ้าย' หรือ 'ขวา' เท่านั้น)")
```
